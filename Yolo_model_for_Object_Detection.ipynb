```python
import os
import cv2
from ultralytics import YOLO

#The best.pt file in YOLO is the saved model checkpoint that contains
# the weights of the best-performing version of the trained model,
#  typically based on validation metrics. This file is used for making
#   predictions on new images and can also be utilized for further
#   fine-tuning or resuming training, ensuring optimal performance
#    in object detection tasks.

# Path to Best.pt
model = YOLO('/content/gdrive/MyDrive/cam11-output/weights/best.pt')

# List of directories with test images (in Google Drive)
test_images_dirs = [
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/100EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/101EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/102EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/103EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/104EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/105EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/106EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL1_301222/107EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/100EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/101EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/102EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/103EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/104EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/105EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/106EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/107EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/108EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/109EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/110EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/111EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/112EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/113EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/114EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/115EK113',
    '/content/gdrive/MyDrive/camera pictures/cam11/CAM11_DL2_170123/116EK113']



# Defining separate directories to save images based on the number of birds detected in each image


output_dirs = {
    1: '/content/gdrive/MyDrive/cam11-output/01',
    2: '/content/gdrive/MyDrive/cam11-output/02',
    3: '/content/gdrive/MyDrive/cam11-output/03',
    4: '/content/gdrive/MyDrive/cam11-output/04',
    5: '/content/gdrive/MyDrive/cam11-output/05'
}

# Ensure output directories exist
for path in output_dirs.values():
    os.makedirs(path, exist_ok=True)

# Class ID for 'Snow Petrel' is set to 0 based on personal preference
snow_petrel_class_id = 0

# Process each directory
for dir_path in test_images_dirs:
    for image_name in os.listdir(dir_path):
        if image_name.lower().endswith(".jpg"):
            image_path = os.path.join(dir_path, image_name)
            img = cv2.imread(image_path)

            if img is None:
                print(f"Failed to load image: {image_path}")
                continue

            # Predict and detect objects
            results = model(img, conf=0.75)

            # Print results to understand the structure
            print(f"Results for {image_name}: {results}")

            # Assuming results is a list and we check for empty detection
            if len(results) > 0 and results[0].boxes is not None:
                detections = results[0].boxes

                # Count Snow Petrel detections
                bird_count = sum(1 for det in detections if int(det.cls) == snow_petrel_class_id)

                # Determine the appropriate output directory
                if bird_count in output_dirs:
                    output_path = os.path.join(output_dirs[bird_count], image_name)
                else:
                    output_path = os.path.join(output_dirs[5], image_name)
            else:
                bird_count = 0

            # Save image to the appropriate directory if there are detections
            if bird_count > 0:
                cv2.imwrite(output_path, img)
            else:
                print(f"No detections for {image_name}")

```
