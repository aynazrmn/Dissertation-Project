```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/gdrive/MyDrive/combined_number_of_birds.csv')


df['Datetime'] = pd.to_datetime(df['Dates'] + ' ' + df['Times'], errors='coerce')
df_sorted = df.sort_values(by=['Camera_Numbers', 'Datetime'])
df_sorted['Time_Diff'] = df_sorted.groupby('Camera_Numbers')['Datetime'].diff().dt.total_seconds()

filtered_groups = []
for camera, group in df_sorted.groupby('Camera_Numbers'):
    indices_to_keep = []
    previous_index = None

    for index, row in group.iterrows():
        if previous_index is not None:
            if pd.notna(row['Time_Diff']) and row['Time_Diff'] < 60:  # Time difference less than 1 minute
                if row['Bird_Number'] > group.loc[previous_index, 'Bird_Number']:
                    indices_to_keep.pop()
                    indices_to_keep.append(index)
                elif row['Bird_Number'] == group.loc[previous_index, 'Bird_Number']:
                    indices_to_keep.pop()
                    indices_to_keep.append(index)
            else:
                indices_to_keep.append(index)
        else:
            indices_to_keep.append(index)

        previous_index = index
    filtered_groups.append(group.loc[indices_to_keep])


filtered_df = pd.concat(filtered_groups)
filtered_df = filtered_df.drop(columns=['Datetime', 'Time_Diff'])
output_path = '/content/gdrive/MyDrive/filtered_data.csv'
filtered_df.to_csv(output_path, index=False)
print(filtered_df)
```
